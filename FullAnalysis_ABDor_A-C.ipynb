# Definition of the astrometric model

In this section we define the needed functions to build the model that describes the astrometry of the binary system. It is based on the definitions of WH09, but we describe below in detail the part that is used in this (more restricted) model.

## Model for the motion of the primary star

### Description of the model

#### Coordinates

All the calculations in this analysis are done based on a particular system of coordinates defined for the particular system studied. These coordinates refer to displacements along rectilinear axes in the ($\alpha$, $\delta$) directions at the nominal position of the star. These are usually refered to as "$\Delta \alpha \cos \delta$" and "$\Delta \delta$", but I will use here the notation "$\alpha_r$" and "$\delta_r$" (for "relative"). 

Given a set of observed positions $(\alpha, \delta)$, these relative coordinates are obtained as

$$
\alpha_r = (\alpha - \alpha_N) \cdot \cos \delta_N
$$
$$
\delta_r = \delta - \delta_N
$$

where ($\alpha_N, $\delta_N$) are the "nominal" coordinates of the star that are kept fixed.

Therefore, what our model should give us (in order to compare to the data) is a prediction, given the parameter values, of the expected position of the star, in terms of these coordinates $\alpha_r$, $\delta_r$, as function of time.

We will break up the model prediction in several "components" corresponding to:
* orbit (due to the presence of the secondary)
* proper motion
* parallax

So, explicitly, the prediction given by the model for the position of the primary star as function of time can be expressed as

$$
\alpha_{r, model}(t) = \alpha_{r, orbit}(t) + \alpha_{r, pm}(t) + \alpha_{r, par}(t) - \alpha_{r, ref}
$$
$$
\delta_{r, model}(t) = \delta_{r, orbit}(t) + \delta_{r, pm}(t) + \delta_{r, par}(t) - \delta_{r, ref}
$$

where ($\alpha_{r, ref}$, $\delta_{r, ref}$) are two parameters of the model that correspond to the position of the center of mass of the system at a pre-fixed "reference time" $T_{ref}$, if there were no parallax.

We give the expressions for the rest of the components below.

#### Orbit

The "orbit" contribution to the motion of the primary star gives the position of the star as function of time with respect to the center of mass of the system. It is computed following the "Thiele-Innes" formalism, and is given by

$$\alpha_{r, orbit}(t) = B \cdot X(t) + G \cdot Y(t)$$
$$\delta_{r, orbit}(t) = A \cdot X(t) + F \cdot Y(t)$$



A, B, F, G are the Thiele-Innes parameters given by 
$$A = a(\cos \Omega \cos \omega - \sin \Omega \sin \omega \cos i)$$
$$B = a(\sin \Omega \cos \omega + \cos \Omega \sin \omega \cos i)$$
$$F = a(- \cos \Omega \sin \omega - \sin \Omega \cos \omega \cos i)$$
$$G = a(- \sin \Omega \sin \omega + \cos \Omega \cos \omega \cos i)$$
where $a, \Omega, \omega_{\star}, i$ are orbital parameters of the model:
* $a$: semi-major axis of the primary star apparent orbit in the sky, in units of arc (here, we use arcsec)
* $\Omega$: longitude of the ascending (approaching) node [angle between 0 and 180 deg]
* $\omega$: longitude of periastron of the orbit [angle between -180 and 180 deg]
* $i$: inclination of the orbit on the sky [angle between 0 and 90 deg, where $i=0$ corresponds to face-on]

$X(t)$, $Y(t)$  are the "elliptical rectangular coordinates", defined as
$$X(t) = \cos E(t) - e$$
$$Y(t) = \sqrt{1 - e^2} \sin E(t)$$
where $e$ is the ellipticity of the orbit (a model parameter) and $E(t)$ is the eccentric anomaly, that is obtained by numerically solving Kepler's equation:
$$E(t) - e \sin E(t) = \frac{2 \pi (t - T_0)}{P}$$
where $T_0$ and $P$ are two additional model parameters corresponding, respectively, to the time of periastron passage and to the period of the orbit (times are given in this work always in years).

#### Proper motion

We use a linear model to describe the proper motion of the system as
$$\alpha_{r, pm}(t) = \mu_{\alpha} (t - T_{ref})$$
$$\delta_{r, pm}(t) = \mu_{\delta} (t - T_{ref})$$

where $\mu_{\alpha}$, $\mu_{\delta}$ are two model parameters describing the proper motion of the system in each direction

#### Parallax

The parallax term is given by
$$\alpha_{r, par}(t) = \pi_{par} \cdot \Pi_{\alpha}(t)$$
$$\delta_{r, par}(t) = \pi_{par} \cdot \Pi_{\delta}(t)$$

where $\pi_{par}$ is the parallax of the system in arcseconds (a model parameter), and $\Pi_{\alpha}(t)$, $\Pi_{\delta}(t)$ are the astrometric displacements due to parallax in the *nominal* direction of the system at the times of observation, and are given by
$$\Pi_{\alpha}(t) = r_x(t) \sin \alpha_N - r_y(t) \cos \alpha_N$$
$$\Pi_{\delta}(t) = \left( r_x(t) \cos \alpha_N + r_y(t) \sin \alpha_N \right) \sin \delta_N - r_z(t) \cos \delta_N$$

and $(r_x, r_y, r_z)(t)$ are the Cartesian components in equatiorial coordinates (in AU) of the position of the Earth with respect to the solar system barycenter at the times of observation (these are taken from the JPL ephemerides, using the Astropy/Pyephem libraries).

### Functions defining the model

```python
def get_ThieleInnes_from_params(a, angle_O_deg, angle_w_deg, angle_i_deg):
    
    sin_O = np.sin(angle_O_deg*DEG2RAD)
    cos_O = np.cos(angle_O_deg*DEG2RAD)
    sin_w = np.sin(angle_w_deg*DEG2RAD)
    cos_w = np.cos(angle_w_deg*DEG2RAD)
    cos_i = np.cos(angle_i_deg*DEG2RAD)
    
    TIpar_A = a*( (cos_O*cos_w) - (sin_O*sin_w*cos_i) )
    TIpar_B = a*( (sin_O*cos_w) + (cos_O*sin_w*cos_i) )
    TIpar_F = a*( (-cos_O*sin_w) - (sin_O*cos_w*cos_i) )
    TIpar_G = a*( (-sin_O*sin_w) + (cos_O*cos_w*cos_i) )
    
    return TIpar_A, TIpar_B, TIpar_F, TIpar_G

def get_XY_from_params(eccentricity, ecc_anomaly):
    X = np.cos(ecc_anomaly) - eccentricity
    Y = np.sqrt(1 - (eccentricity**2))*np.sin(ecc_anomaly)
    return X, Y

def f_Kepler(x, eccentricity, M):
    return x - eccentricity*np.sin(x) - M

def get_ecc_anomaly(tval, eccentricity, period, T0):
    
    cycle_no = np.floor( (tval - T0) / period)
    low_lim = 2.*np.pi*cycle_no
    upp_lim = 2.*np.pi*(cycle_no + 1)
    M_t = 2.*np.pi*(tval - T0) / period
    
    return so.brentq(f=f_Kepler, args=(eccentricity, M_t), a=low_lim, b=upp_lim)

def get_XY_from_time(tval, eccentricity, period, T0):
    ecc_anom = get_ecc_anomaly(tval, eccentricity, period, T0)
    return get_XY_from_params(eccentricity, ecc_anom)


def compute_orbit_times(tvals, angle_O_deg, angle_w_deg, angle_i_deg, a_axis,
                        eccentricity, period, T0):
    
    tvals = np.atleast_1d(tvals)
    Nt = len(tvals)
    orbit_vals_delta_alphacosd = np.empty((2, Nt))
    
    A, B, F, G = get_ThieleInnes_from_params(a_axis, angle_O_deg, angle_w_deg,
                                             angle_i_deg)
    
    for i,t in enumerate(tvals):
        X, Y = get_XY_from_time(t, eccentricity, period, T0)
        orbit_vals_delta_alphacosd[0, i] = A*X + F*Y
        orbit_vals_delta_alphacosd[1, i] = B*X + G*Y
        
    return orbit_vals_delta_alphacosd
```

```python
def compute_proper_motions(tvals, mu_alpha, mu_delta, t_ref):
    pm_delta = mu_delta*(tvals - t_ref)
    pm_alpha = mu_alpha*(tvals - t_ref)
    return np.array((pm_delta, pm_alpha))

def get_earth_position_time(time_years):
    t = Time(time_years, format='decimalyear')
    with coord.solar_system_ephemeris.set('de430'):
        cart_repr_earth = coord.get_body_barycentric('earth', time=t)
    return cart_repr_earth.xyz.to(u.AU).value

def get_parallax_factors(time_years, skycoords_ref):
    earth_coords = get_earth_position_time(time_years)
    
    sin_ra = np.sin(skycoords_ref.ra)
    cos_ra = np.cos(skycoords_ref.ra)
    sin_dec = np.sin(skycoords_ref.dec)
    cos_dec = np.cos(skycoords_ref.dec)
    
    Par_factor_ra = earth_coords[0]*sin_ra - earth_coords[1]*cos_ra
    Par_factor_dec = ( earth_coords[0]*cos_ra + earth_coords[1]*sin_ra )*sin_dec - earth_coords[2]*cos_dec
    
    return np.array([Par_factor_dec, Par_factor_ra])

def compute_parallax(pi_parallax_param, parallax_factors):
    return pi_parallax_param*parallax_factors
```

```python
def mean_dec_ra_predicted(tvals, angle_O_deg, angle_w_deg, angle_i_deg, a_axis,
                          eccentricity, period, T0, mu_delta, mu_alpha,
                          pi_parallax, t_ref, parallax_factors):
    
    orbit_term = compute_orbit_times(tvals, angle_O_deg, angle_w_deg,
                                     angle_i_deg, a_axis, eccentricity, period,
                                     T0)
    proper_motion_term = compute_proper_motions(tvals, mu_alpha, mu_delta, t_ref)
    parallax_term = compute_parallax(pi_parallax, parallax_factors)
    
    return orbit_term + proper_motion_term + parallax_term
```

## Model for the relative (secondary) orbit

### Description of the model

When considering the motion of the secondary star, the reference, proper motion and parallax terms are identical to that of the primary star. The only difference is in the orbit component, which follows the same equations, but with two changes:

* The semi-major axis of the orbit of the secondary, $a_{sec}$, differs of that of the primary by a factor $q$ defined as the ratio of the masses:
$$a_{sec} = \frac{a_{prim}}{q_m} \;\; ; \;\; q_m = \frac{m_{sec}}{m_{prim}}$$

* The longitude of periastron changes by a value of $\pi$ (180 deg):
$$\omega_{sec} = \omega_{prim} + \pi$$

Therefore, the only change will be in the definition of the Thiele-Innes parameters. If we express everything in terms of the parameters of the *primary* star (which will be the parameters for which we perform the fit), these will be given, for the secondary, as (note the sign changes)
$$A_{sec} = \frac{a}{q_m}(-\cos \Omega \cos \omega + \sin \Omega \sin \omega \cos i)$$
$$B_{sec} = \frac{a}{q_m}(-\sin \Omega \cos \omega - \cos \Omega \sin \omega \cos i)$$
$$F_{sec} = \frac{a}{q_m}(\cos \Omega \sin \omega + \sin \Omega \cos \omega \cos i)$$
$$G_{sec} = \frac{a}{q_m}(\sin \Omega \sin \omega - \cos \Omega \cos \omega \cos i)$$

In the case of ABDor A/C, we do not have observations of the secondary's absolute orbit, only of the relative orbit between the primary and the secondary. In this case, the prediction of the model is simply the difference between the predicted coordinates for both stars (with the appropriate sign to match that used to report the observations).


### Functions defining the model

```python
def get_ThieleInnes_from_params_secondary(a_primary, angle_O_deg,
                                          angle_w_deg_prim, angle_i_deg,
                                          q_massratio):
    
    a = a_primary/q_massratio
    
    sin_O = np.sin(angle_O_deg*DEG2RAD)
    cos_O = np.cos(angle_O_deg*DEG2RAD)
    sin_w = -np.sin(angle_w_deg_prim*DEG2RAD) # w_sec = w_prim + pi
    cos_w = -np.cos(angle_w_deg_prim*DEG2RAD) # w_sec = w_prim + pi
    cos_i = np.cos(angle_i_deg*DEG2RAD)
    
    TIpar_A = a*( (cos_O*cos_w) - (sin_O*sin_w*cos_i) )
    TIpar_B = a*( (sin_O*cos_w) + (cos_O*sin_w*cos_i) )
    TIpar_F = a*( (-cos_O*sin_w) - (sin_O*cos_w*cos_i) )
    TIpar_G = a*( (-sin_O*sin_w) + (cos_O*cos_w*cos_i) )
    
    return TIpar_A, TIpar_B, TIpar_F, TIpar_G


def compute_relative_orbit_times(tvals, angle_O_deg, angle_w_deg, angle_i_deg,
                                 a_axis, eccentricity, period, T0, q_massratio):
    # Note w, a in the input parameters refer to those of the *primary*
    
    tvals = np.atleast_1d(tvals)
    Nt = len(tvals)
    relorbit_vals_delta_alphacosd = np.empty((2, Nt))
    
    # Primary
    A, B, F, G = get_ThieleInnes_from_params(a_axis, angle_O_deg, angle_w_deg,
                                             angle_i_deg)
    # Secondary
    A_sec, B_sec, F_sec, G_sec = \
        get_ThieleInnes_from_params_secondary(a_axis, angle_O_deg, angle_w_deg,
                                              angle_i_deg, q_massratio)
        
    for i,t in enumerate(tvals):
        X, Y = get_XY_from_time(t, eccentricity, period, T0)
        relorbit_vals_delta_alphacosd[0, i] = A*X + F*Y - (A_sec*X + F_sec*Y)
        relorbit_vals_delta_alphacosd[1, i] = B*X + G*Y - (B_sec*X + G_sec*Y)
        
    return relorbit_vals_delta_alphacosd
```

## Summary: parameters of the model

In summary, the model has 13 free parameters that we will fit for (they are only 12 in the case in which we only have observations of the primary orbit):

* $\Omega$: longitude of the ascending (approaching) node [angle between 0 and 180 deg]
* $\omega$: longitude of periastron of the orbit of the *primary* star[angle between -180 and 180 deg]
* $i$: inclination of the orbit on the sky [angle between 0 and 90 deg]
* $a$: semi-major axis of the *primary* star apparent orbit in the sky [arcsec]
* $e$: eccentricity of the orbit
* $P$: period of the orbit [in years]
* $T_0$: time of periastron passage [in years, using the convention that this is the first periastron passage after $T_{ref}$]
* $\mu_{\alpha}$: proper motion in the RA direction [arcsec / year]
* $\mu_{\delta}$: proper motion in the DEC direction [arcsec / year]
* $\pi_{par}$: parallax of the system [in arcsec]
* $\delta_{r, ref}$: DEC coordinate of the position of the center of mass of the system, without parallax, at the reference time $T_{ref}$ [arcsec]
* $\alpha_{r, ref}$: RA coordinate of the position of the center of mass of the system, without parallax, at the reference time $T_{ref}$ [arcsec]
* $q_m$: ratio of the mass of the secondary to that of the primary. This parameter can not be constrained in the case where we only have observations of the primary star orbit

# Definition of the "statistical model" for `emcee`

For the MCMC calculation to be done by `emcee` we need to define three functions (see docs in http://dfm.io/emcee/current/ ):
* the (log of the) prior of the parameters
* the log-likelihood of the observed data given the parameters
* the log of the posterior probability of the parameters given the data (this is just the sum of the two above)

## Definition of the prior

The prior of the parameters is quite hard-coded here. I assume (mostly) independent prior probabilities for each of the parameters. I use a combination of uniform, normal and half-normal (i.e., restricted to the positive part) priors depending on the parameter, see the exact definition in the code. The widths of the different priors are hard coded here, but can be changed.

NOTES:
* The priors of $T_0$ and $P$ are actually not independent, as the prior on $T_0$ is a uniform distribution between $T_{ref}$ and $T_{ref} + P$
* If you change the definition of the prior here (e.g. the widhts), remember that you should change it also in the part where we generate the initial positions for the walkers

```python
def lnprior(theta, ref_time):
    angle_O, angle_w, angle_i, a_axis, eccentricity, period, \
        T0, mu_alpha, mu_delta, pi_parallax, dec_r_ref, alpha_r_ref, q_mass = theta
        
    lprior = 0
    lprior += st.uniform.logpdf(angle_O, loc=0, scale=180)
    lprior += st.uniform.logpdf(angle_w, loc=-180, scale=360)
    lprior += st.uniform.logpdf(angle_i, loc=0, scale=90)
    lprior += st.halfnorm.logpdf(a_axis, scale=0.1)
    lprior += st.uniform.logpdf(eccentricity, loc=0, scale=1)
    lprior += st.uniform.logpdf(period, loc=0, scale=25)
    lprior += st.uniform.logpdf(T0, loc=ref_time, scale=period)
    lprior += st.norm.logpdf(mu_alpha, loc=0, scale=0.3)
    lprior += st.norm.logpdf(mu_delta, loc=0, scale=0.3)
    lprior += st.halfnorm.logpdf(pi_parallax, scale=0.1)
    lprior += st.norm.logpdf(dec_r_ref, loc=0, scale=0.5)
    lprior += st.norm.logpdf(alpha_r_ref, loc=0, scale=0.5)
    lprior += st.halfnorm.logpdf(q_mass, scale=1.0)
    
    return lprior
```

## Definition of the likelihood

We compute a simple likelihood for the data, combining both the data for the absolute orbit of the primary and for the relative orbit between the primary and the secondary, using the two simple assumptions:
* The errors on the observed coordinates are normally distributed
* The errors in the RA and DEC directions are independent (i.e., they are not correlated)

Under these assumptions, the log-likelihood is basically a $\chi^2$ statistic.

```python
def lnlike(theta, t, data_dec_vals, data_ra_vals, data_dec_err, data_ra_err,
           parallax_factors_t,
           t_rel, data_rel_dec, data_rel_ra, data_rel_dec_err, data_rel_ra_err,
           ref_time):
    angle_O, angle_w, angle_i, a_axis, eccentricity, period, \
        T0, mu_alpha, mu_delta, pi_parallax, Ddec_ref, Dracd_ref, q_mass = theta
        
    delta_decra_model_noref = mean_dec_ra_predicted(t, angle_O_deg=angle_O,
                                              angle_w_deg=angle_w, angle_i_deg=angle_i,
                                              a_axis=a_axis, eccentricity=eccentricity,
                                              period=period, T0=T0,
                                              mu_delta=mu_delta, mu_alpha=mu_alpha,
                                              pi_parallax=pi_parallax, t_ref=ref_time,
                                              parallax_factors=parallax_factors_t)
    
    delta_decra_model = (delta_decra_model_noref.T - np.array((Ddec_ref, Dracd_ref))).T
    
    xi2_dec = (((delta_decra_model[0] - data_dec_vals)**2)/(data_dec_err**2)).sum()
    xi2_ra = (((delta_decra_model[1] - data_ra_vals)**2)/(data_ra_err**2)).sum()
    
    # Relative orbit
    delta_decra_rel_model = \
        compute_relative_orbit_times(tvals=t_rel, angle_O_deg=angle_O, 
                                     angle_w_deg=angle_w, angle_i_deg=angle_i,
                                     a_axis=a_axis, eccentricity=eccentricity,
                                     period=period, T0=T0, q_massratio=q_mass)
    
    xi2_dec_rel = (((delta_decra_rel_model[0] - data_rel_dec)**2)/(data_rel_dec_err**2)).sum()
    xi2_ra_rel = (((delta_decra_rel_model[1] - data_rel_ra)**2)/(data_rel_ra_err**2)).sum()
    
    return -0.5*(xi2_dec + xi2_ra + xi2_dec_rel + xi2_ra_rel)
```

## Definition of the posterior

Just add the two prefious terms together, with some care for +/- infinite values

```python
def lnpostprob(theta, t, data_dec_vals, data_ra_vals, data_dec_err, data_ra_err,
               parallax_factors_t,
               t_rel, data_rel_dec, data_rel_ra, data_rel_dec_err, data_rel_ra_err,
               ref_time):
    lp = lnprior(theta, ref_time)
    
    ## REALLY need to do this to avoid problems later!
    if not np.isfinite(lp):
        return -np.inf
    return lp + lnlike(theta, t, data_dec_vals, data_ra_vals, data_dec_err,
                       data_ra_err, parallax_factors_t, t_rel, data_rel_dec, data_rel_ra,
                       data_rel_dec_err, data_rel_ra_err, ref_time)
```

```python
angle_O_init = st.uniform.rvs(loc=0, scale=180, size=(ntemps, nwalkers))
angle_w_init = st.uniform.rvs(loc=-180, scale=360, size=(ntemps, nwalkers))
angle_i_init = st.uniform.rvs(loc=0, scale=90, size=(ntemps, nwalkers))
a_axis_init = st.halfnorm.rvs(loc=0, scale=0.1, size=(ntemps, nwalkers))
eccent_init = st.uniform.rvs(loc=0, scale=1, size=(ntemps, nwalkers))
period_init = st.uniform.rvs(loc=0, scale=20, size=(ntemps, nwalkers))
T0_init = np.array([st.uniform.rvs(loc=ref_time, scale=p, size=1) for p in period_init.flatten()]).reshape((ntemps, nwalkers))
mu_alpha_init = st.norm.rvs(loc=0, scale=0.3, size=(ntemps, nwalkers))
mu_delta_init = st.norm.rvs(loc=0, scale=0.3, size=(ntemps, nwalkers))
pi_par_init = st.halfnorm.rvs(loc=0, scale=0.1, size=(ntemps, nwalkers))
dec_ref_init = st.norm.rvs(loc=0, scale=0.5, size=(ntemps, nwalkers))
alpha_ref_init = st.norm.rvs(loc=0, scale=0.5, size=(ntemps, nwalkers))
q_mass_init = st.halfnorm.rvs(loc=0, scale=1.0, size=(ntemps, nwalkers))
```

## Derive masses of the primary and secondary stars from the model parameters

The masses of the primary and the secondary stars ($m_p$, $m_s$) can be obtained from an appropriate combination of the model parameters (using Kepler's third law). 
In order to obtain accurately the posterior PDF for the masses' values, we will derive them using the full chains (after burn in). In this way, e.g., we take correctly into account the covariance between model parameters when we propagate the errors to obtain the constraints on the masses.

From eq. (55) in WH09 and the definition of $q_m$, we obtain:

* For the primary mass:
$$m_p = \frac{a^3 (q_m + 1)^2}{q_m^3 \pi_{par}^3 P^2}$$

* For the secondary mass:
$$m_s = \frac{a^3 (q_m + 1)^2}{q_m^2 \pi_{par}^3 P^2} = q_m m_p$$

For clarity, I show here a (reduced) corner plot, considering only the (physical) parameters that affect this calculation.

```python
interesting_parameters = [r'$a$', r'$P$', r'$\pi_{par}$', r'$q_m$']
fig = c.plotter.plot(parameters=interesting_parameters)
```

```python
def mass_primary_from_params(q, a, pi, P):
    return (a**3)*((1+q)**2) / ( (q**3)*(pi**3) * (P**2))

def mass_secondary_from_params(q, a, pi, P):
    return q*mass_primary_from_params(q, a, pi, P)
```

## Best fit and comparison to data

In order to report the constraints on the parameters from a Bayesian analysis, we don't need to get a "best fit" model, but we do this here for completeness, and to decide which model to use for the comparison plots with the data.

We define the "best-fit" model as the "maximum a posteriori" (MAP) model, i.e. the point in parameter space where the posterior PDF has a maximum.

### Obtain the MAP "best-fit" model

We obtain the MAP using a two-step process:
1. We find the sample in the chain with the maximum value of the log-posterior
2. We now use a minimization procedure to find the MAP, using as a starting point the sample found in the previous step

```python
print(lnprob_T1.max())
walker_lpmax = lnprob_T1.max(axis=1).argmax()
iter_lpmax = lnprob_T1[walker_lpmax].argmax()
print(lnprob_T1[walker_lpmax, iter_lpmax])
print(walker_lpmax, iter_lpmax)
print(lnprob_T1[walker_lpmax, iter_lpmax])
maxpost_from_chain_pos = chain_T1[walker_lpmax, iter_lpmax]
```

```python
nlp = lambda *args: -lnpostprob(*args)
result = so.minimize(nlp, maxpost_from_chain_pos, args=(obs_times, rel_dec_obs_arcsec,
                                       rel_racd_obs_arcsec, obs_err_dec_arcsec,
                                       obs_err_racd_arcsec, pfactors_obst,
                                       obs_times_rel, dec_vals_rel, ra_vals_rel,
                                       dec_vals_err_rel, ra_vals_err_rel,
                                       ref_time),
                    method='Powell')
print(result.message)
```

To have a rough idea of the "goodness of fit", we use the $\chi^2$ statistic:

```python
chi2_min_from_Powell = -2.*lnlike(result['x'], obs_times, rel_dec_obs_arcsec,
                                       rel_racd_obs_arcsec, obs_err_dec_arcsec,
                                       obs_err_racd_arcsec, pfactors_obst,
                                       obs_times_rel,
                                       dec_vals_rel, ra_vals_rel,
                                       dec_vals_err_rel, ra_vals_err_rel,
                                       ref_time)
```

```python
n_dof = (2*size_total) - ndim
print("We obtain a \chi^2 = {:.3f} for {:d} degrees of freedom, "
      "resulting in a reduced \chi^2_r = {:.3f}".format(chi2_min_from_Powell, n_dof, chi2_min_from_Powell/n_dof))
```
